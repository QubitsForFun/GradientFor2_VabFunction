```python
# z = f(x,y) = x^1.1 - sin(x + 1.3*y) Буду брать частные производные по x и y используя GradientTape

import tensorflow as tf
import numpy as np
import matplotlib as plt
import pandas as pd
```

```python
x_values = np.linspace(0.0, 4.0, 41) #На основании этого буду строить сетку x-y , в её узлах будут значения dz/dx(x,y) и dz/dy(x,y)
y_values = np.linspace(0.0, 4.0, 41)

X_grid , Y_grid = np.meshgrid(x_values, y_values)

x = tf.Variable(X_grid)
y = tf.Variable(Y_grid)
```

```python
with tf.GradientTape(persistent = True) as tape: #persistent = True позволит мне сразу брать как производные по x, так и по y
  tape.watch(x)
  tape.watch(y)
  z = x ** 1.1 - tf.sin(x + 1.3 * y)
```

```python
dz_dx, dz_dy = tape.gradient(z, [x, y])
```

```python
#Для удобства конвертирую в формат numpy, заметил на это оч. много времени уходит, далее надо попробовать что-то более низкоуровневое, или вообще сделать без numpy
dz_dx_np = dz_dx.numpy()
```

```python
df_dz_dx = pd.DataFrame(dz_dx_np, columns = y_values, index = x_values) #по идее, должны они наоборот выводиться, если сравнивать с excel-файлом, где я от руки считал
print(df_dz_dx)
```

```python
dz_dy_np = dz_dy.numpy()
df_dz_dy = pd.DataFrame(dz_dx_np, columns = y_values, index = x_values)
print(df_dz_dy)
```
